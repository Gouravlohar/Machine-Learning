```python
import seaborn as sns
```

# Bivariate

## 1.Scatterplot(Numerical-Numerical)

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
```

# Multivariate

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'])
```

# 2.Bar Plot(Numerical-Categorical)

### The rod like line is ***confidence Interval***<br>
A confidence interval, in statistics, refers to the probability that a population parameter will fall between a set of values for a certain proportion of times

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Fare'])
```

### Multivariate

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Fare'],hue=titanic['Sex'])
```

# 3. Box Plot(Numerical-Categorical)

```python
sns.boxplot(x=titanic['Pclass'],y=titanic['Fare'])
```

# Distplot(Numerical-Categorical)

```python
sns.distplot(x=titanic['Age'])
```

## Multivariate

```python
sns.distplot(x=titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(x=titanic[titanic['Survived']==1]['Age'],hist=False)
```

# HeatMap(Categorical-Categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass').mean()['Survived']*100
```

# 6.CLusterMap(Categorical-Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
